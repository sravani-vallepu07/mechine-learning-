```python
# heatmap
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
gb=np.linspace(1,10,20).reshape(4,5)
sns.heatmap(gb,vmin=0,vmax=10,cmap='gist_heat')
```

```python
data=sns.load_dataset("anagrams")
x=data.drop(columns=["attnr"],axis=1)
sns.heatmap(x)
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
sns.heatmap(var1,vmin=0,vmax=10,cmap='gist_heat',annot=True)
plt.show()
```

```python
aru=np.array([['a0','a1','a2','a3','a4'],
               ['b0','b1','b2','b3','b4']])
sns.heatmap(var1, vmin=1, vmax=10, cmap='hsv', annot=aru,fmt='s')
```

```python
y={'fontsize':20,"color":"r"}
sns.heatmap(gb,vmin=0,vmax=10,cmap="hsv",annot=True,annot_kws=y,linewidth=10,linecolor='y',cbar=False,xticklabels=False)
```

```python
v=sns.heatmap(gb,vmin=0,vmax=10,cmap="hsv",annot=True,annot_kws=y,linewidth=10,linecolor='y',cbar=False,xticklabels=False)
v.set(xlabel="python",ylabel="wscode")
sns.set(font_scale=0.2)
plt.show()
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

cds = sns.load_dataset("tips")

fg = sns.FacetGrid(cds, col="sex")
fg.map(plt.scatter, "total_bill", 'tip')
plt.show()

```

```python
mk=sns.FacetGrid(cds,col='sex',hue='day')
mk.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
hj=sns.FacetGrid(cds,col="day",hue="sex")
hj.map(plt.bar,"total_bill","tip",edgecolor="r").add_legend()
plt.show()
```
