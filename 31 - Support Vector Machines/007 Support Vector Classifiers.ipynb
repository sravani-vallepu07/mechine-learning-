```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##lets create synthetic data points
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_redundant=0)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0], y=pd.DataFrame(x)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
svc.coef_
```

```python
##prediction
y_pred=svc.predict(x_test) 
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
rbf=SVC(kernel='rbf')
# best accuracy giving kernel rbf
```

```python
rbf.fit(x_train,y_train)
```

```python
y_pred1=rbf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_pred1,y_test))
print(confusion_matrix(y_pred1,y_test))
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(x_train,y_train)
y_pred2=polynomial.predict(x_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_pred2,y_test))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(x_train,y_train)
y_pred3=sigmoid.predict(x_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_pred3,y_test))
```

```python
sigmoid.intercept_
```

## hyperparameter  tuning with VC


```python
from sklearn.model_selection import GridSearchCV
params_grid={ 'C':[0.1,1,10,100,1000],
             'gamma':[1,0.1,0.01,0.001,0.0001],
             'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=params_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```
