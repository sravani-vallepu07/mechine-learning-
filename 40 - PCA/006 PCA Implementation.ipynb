```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
##load the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_datset=load_breast_cancer()
```

```python
cancer_datset.keys()
```

```python
print(cancer_datset.DESCR)
```

```python
df=pd.DataFrame(cancer_datset['data'],columns=cancer_datset['feature_names'])
df.head()
```

```python
###standarzation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data
```

```python
##applying PCA Alogorithms
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
data_pca=pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,8))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_datset['target'],cmap='plasma')
plt.xlabel('first principal component')
plt.ylabel('second principal componenet')
plt.legend()
plt.show()
```
